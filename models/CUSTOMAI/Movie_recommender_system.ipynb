```python
## we have to convert the geners to correct form - list of string like action
## adeventure, sci-fi
```

```python
import ast
```

```python
def preproc(d) :
    l = []
    for i in ast.literal_eval(d) :
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(preproc)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(preproc)
```

```python
movies.head(1)
```

```python
## we only need top 3 cast because they are most important and from crew we
## need only director's information
```

```python
movies.iloc[0].cast
```

```python
def preproc1(d) :
    l = []
    count = 0
    for i in ast.literal_eval(d) :
        if(count != 3) :
            count = count+1
            l.append(i['name'])
        else :
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(preproc1)
```

```python
movies.iloc[0].crew
```

```python
def preproc2(d) :
    l = []
    for i in ast.literal_eval(d) :
        if(i['job'] == 'Director') :
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(preproc2)
```

```python
movies.head()
```

```python
## convert the overview column from string to list
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(1)
```

```python
## we have to remove space between strings because space will make them 
## different entities
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
df = movies[['id','title','tags','popularity','vote_count']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
```

```python
df.head(1)
```

```python
## here is a problem in featuure name, here some word like action and actions
## are diff which is not a good thing to get rid of this thing we will use
## stemming ,it converts ['loved','loving','love'] to ['love','love','love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(text) :
    l = []
    for i in text.split() :
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
```

```python
df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = 'english',max_features = 5000)
```

```python
vect = cv.fit_transform(df['tags']).toarray()
```

```python
vect
```

```python
cv.get_feature_names()
```

```python
## as here dimensions are high(5000) so euclidean distance is not reliable so
## we will calculate cosine distance i.e angle between vecctors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vect)
```

```python
sim[1]
```

```python
## now if a movie is given than show top 5 similar movies
```

```python
def similarmovies(movie) :
    ind = df[df['title'] == movie].index[0]
    recom_list = sorted(list(enumerate(sim[ind])),reverse = True,key = lambda x : x[1])[1:6]
    for i in recom_list :
        print(df.iloc[i[0]].title)
```

```python
similarmovies('Batman Begins')
```
